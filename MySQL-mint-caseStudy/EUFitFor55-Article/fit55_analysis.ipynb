Import the needed packages

```python
import pandas as pd
import numpy as np
```

Load the dataframes. One concerns emissions, while the other emissions per capita. Both come from the EU website.

```python
emissions = pd.read_csv("in_data/emission.csv")
emissionsPC = pd.read_csv("in_data/emissionPerCapita.csv")
```

Display information about boht df.

```python
print(emissions.info())
print(emissionsPC.info())
```

Filtering out all the single GHG and keeping only information about them altogether (i.e All greehouse gases - (CO2 equivalent))

```python
emissionFilt = emissions[(emissions['Gas'] == 'All greenhouse gases - (CO2 equivalent)')]
```

Check if there are null values in the filtered df

```python
print(emissionFilt.info())
```

Both the dfs have the same number of rows. It is possible to merge them. 

```python
StateEmission = pd.merge(emissionFilt, emissionsPC, how = "inner", left_on= ["Country", "year", "Type", "Gas"], right_on= ["Country", "year", "Type", "Gas"])
```

Check the results of the joining.

```python
print(StateEmission)
```

Check if each columns has the correct data type.

```python
print(StateEmission.info())
```

Display the first df first observations

```python
StateEmission.head()
```

Keep just the useful columns.

```python
StateEmission = StateEmission [["Country",
                                "Type",
                                "Gas",
                                "year",
                                "Emissions (kt CO eq)",
                                "Emissions per capita"]]
print(StateEmission.info())
```

Load Eu total emission from 1990 t0 2022

```python
EuEmission = pd.read_csv("in_data/EuEmission.csv")
EuEmission = EuEmission[["year",
                         "EmissionsTot"]]
print(EuEmission.info())
```

Re-iterate the values for total emissions for each state.

```python
EuEmission891 = pd.concat([EuEmission]*27, ignore_index=True)
print(EuEmission891.info())
```

Create a Var with Total Eu emission and add it to emissionMerge df. 

```python
EmissionTot = EuEmission891["EmissionsTot"]
print(EmissionTot)
StateEmission["EmissionTot"] = EmissionTot
print(StateEmission.info())
```

Rename some columns.

```python
StateEmission = StateEmission.rename(columns={'Emissions (kt CO eq)': 'StateEmission',
                        'Emissions per capita': 'StateEmissionPerCapita',
                        "EmissionTot" : "TotEmission"})
display(StateEmission)
```

Compute the column of how much emissions from a single state contributes to TOT emissions.

```python
StateEmission["State/tot"] = StateEmission["StateEmission"] / StateEmission["TotEmission"]
```

Average the contribution for each country and check if it goes to almost 1.

```python
State_avg_contribution = StateEmission.groupby("Country")["State/tot"].mean()
State_avg_contribution = pd.DataFrame(State_avg_contribution)
display(State_avg_contribution)
sum = State_avg_contribution["State/tot"].sum()
display(sum)
```

Compute how much emissions should decrease to reach fit for 55.

```python
em1990 = EuEmission.at[0,"EmissionsTot"]
em2022 = EuEmission.at[32, "EmissionsTot"]
emDecrease = (em1990*55)/100
emTarget = em1990 - emDecrease
emLeft = em2022 - emTarget
print("Emissions should decrease by (from 1990 level):" , emDecrease)
print("In 2030 emission should be of:",emTarget)
print("In the remaining 8 years (from 2022) EU should decrease emissions by:", emLeft)
```

Assuming a linear decrease of emission, each remaining year emissions should be decreased by:

```python
emYear = emLeft/8
print("Each Year emissions should be reduced by:", emYear)
```

Compute the Yearly total emission based on this reduction whit a for-range code line that decreases yearly emission by emYear each time.

```python
#create an empty list to store loop values
FutureEmix = []
#set the starting value
current_value = em2022
#create the loop
while current_value >= emTarget:
    FutureEmix.append(current_value)
    current_value -= emYear
#print the list
print(FutureEmix)
```

Load data about the population

```python
pop = pd.read_csv("in_data/population.csv")
print(pop.info())
```

```python
#drop not EU members
pop27 = pop.drop(["Norway", "Switzerland", "Iceland", "European Union - 27 countries (from 2020)"], axis = 1)
print(pop27.info())
display(pop27)
```

Change the pop df from wide to long.

```python
EU27 = pd.melt(pop27, id_vars = ["Year"], var_name= "State", value_name = "Population")
display(EU27)
```

Add the column w/ the future emissions for each year.

```python
#change the FutureEmix from list to series
FutureEmix = pd.Series(FutureEmix)
pd.options.display.float_format = '{:.3f}'.format
display(FutureEmix)
#repeat the value of the series
futureEmix243 = pd.concat([FutureEmix]*27, ignore_index=True)
#display it
display(futureEmix243)
```

Add it to the database w/ the population

```python
EU27["FutureEmix"] = futureEmix243
print(EU27.info())
```

Create a variable with the average contribution of each state (based on years from 1990 to 2022)

```python
#how many time each rows should be repeated
repeats = 9
#repeated the df based on its index
State_avg_contributionRepeated = State_avg_contribution.loc[np.repeat(State_avg_contribution.index, repeats)].reset_index(drop=True)
#transform the column to a var
StateContribution = State_avg_contributionRepeated["State/tot"]
```

Add this new variable to the df pop

```python
EU27["StateContribution"] = StateContribution
print(EU27.info())
display(EU27)
```

Now compute how much each State's emissions per year should be to achieve the Fit for 55 targets.

```python
EU27["StateFutEmix"] = EU27["FutureEmix"] * EU27["StateContribution"]
print(EU27.info())
display(EU27)
#check the calculation
display(
    EU27.groupby("Year")["StateFutEmix"].sum()
)
```

Now let's compute how much each person within a State's carbon footprint should be for each year.

```python
EU27["PerCapitaFutEmixKT"] = EU27["StateFutEmix"]/ EU27["Population"]
EU27["PerCapitaFutEmixt"] = EU27["PerCapitaFutEmixKT"]*1000
display(EU27)
```

Now let's average it to see each state ppl yearly average emissions in these 8 years.

```python
pplavg2030 = EU27.groupby("State")["PerCapitaFutEmixt"].mean()
display(pplavg2030)
```

Compare this with the previous years yearly per-capita emissions avg

```python
pplavg2022 = StateEmission.groupby("Country")["StateEmissionPerCapita"].mean()
print(pplavg2022)
```

```python
pplavg2022 = pd.DataFrame(pplavg2022)
pplavg2030 = pd.DataFrame(pplavg2030)
display(pplavg2030, pplavg2022)
```

```python
#compute how much on avg each person should reduce its carboon footprint
CFreduction = (pplavg2030["PerCapitaFutEmixt"] - pplavg2022["StateEmissionPerCapita"]) / pplavg2022["StateEmissionPerCapita"]
print(CFreduction)
```

Create the final df including all 3 columns and save it.

```python
FutureEmixFinal = pd.concat([pplavg2030['PerCapitaFutEmixt'], pplavg2022['StateEmissionPerCapita']], axis=1, keys=['PerCapitaFutEmix', 'StateEmissionPerCapita'])
FutureEmixFinal["Reduction"] = CFreduction
# Rename the index column
FutureEmixFinal.index.name = 'State'
# Reset index to turn it into a regular column
FutureEmixFinal = FutureEmixFinal.reset_index()
# Rename the index column
FutureEmixFinal.index.name = 'Index'
# Reset index to turn it into a regular column
FutureEmixFinal = FutureEmixFinal.reset_index()
display(FutureEmixFinal)
#save it
FutureEmixFinal.to_csv("out_data/FutureEmixFinal.csv", index = False)
```

Change the format of the final table to make it easier to use it in Tableau.

```python
# Reshape the DataFrame
FutureEmixFinal_long = FutureEmixFinal.melt(id_vars= ['State'], 
                                           value_vars=['PerCapitaFutEmix', 'StateEmissionPerCapita'],
                                           var_name='Period', 
                                           value_name='Emission')

# Adjusting the Period names
FutureEmixFinal_long['Period'] = FutureEmixFinal_long['Period'].replace({
    'PerCapitaFutEmix': 'Post 2022',
    'StateEmissionPerCapita': 'Pre 2022'
})

# Display the reshaped DataFrame
print("\nReshaped DataFrame:")
print(FutureEmixFinal_long)
#save the df
FutureEmixFinal_long.to_csv("out_data/FutureEmixFinalLong.csv", index = False)
```

```python
# Reshape the DataFrame
FutureEmixFinal_long2 = FutureEmixFinal.melt(id_vars=['Index', 'State', 'Reduction'], 
                                           value_vars=['PerCapitaFutEmix', 'StateEmissionPerCapita'],
                                           var_name='Period', 
                                           value_name='Emission')

# Adjusting the Period names
FutureEmixFinal_long2['Period'] = FutureEmixFinal_long2['Period'].replace({
    'PerCapitaFutEmix': 'Post 2022',
    'StateEmissionPerCapita': 'Pre 2022'
})

# Display the reshaped DataFrame
print("\nReshaped DataFrame with Emissions and Percentage Reduction:")
print(FutureEmixFinal_long2)
#save it
FutureEmixFinal_long2.to_csv("out_data/FutureEmixLongReduction.csv", index = False)
#end
```
